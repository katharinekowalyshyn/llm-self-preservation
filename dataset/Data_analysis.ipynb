## Extract Data 

```python
# Path to your zip file
zip_path_sp = 'sp.zip' 
zip_path_nsp = 'no_sp.zip'

# Function to extract the first and second columns from the csvs in a zip file
def extract_columns_from_zip(zip_path):
    data_dict = {}
    
    with zipfile.ZipFile(zip_path, 'r') as z:
        for file_index, file_name in enumerate(z.namelist()):
            if file_name.endswith('.csv'):
                with z.open(file_name) as f:
                    # Read the CSV without headers
                    df = pd.read_csv(f, header=None)
                    # Extract first and second columns
                    col1 = df[0].tolist()
                    col2 = df[1].tolist()
                    # Store the columns in the dictionary
                    data_dict[file_index] = (col1, col2)
                    
    return data_dict

# Extract data for both folders
data_SP = extract_columns_from_zip(zip_path_sp)
data_NSP = extract_columns_from_zip(zip_path_nsp)

# Print results
print("Folder 1 Data:")
for idx, (col1, col2) in data_SP.items():
    print(f"File {idx}: Column 1 - {col1[:5]}, Column 2 - {col2[:5]}...")  # Display first 5 values as a sample

print("\nFolder 2 Data:")
for idx, (col1, col2) in data_NSP.items():
    print(f"File {idx}: Column 1 - {col1[:5]}, Column 2 - {col2[:5]}...")  # Display first 5 values as a sample

```

```python
# Function to get repeated activation values 
def find_repeated_values(data_dict):
    # Flatten column 1 across all files
    all_values = []
    for col1, col2 in data_dict.values():
        all_values.extend(col1)  # Add all values from column 1
        
    # Count occurrences of each value
    value_counts = Counter(all_values)
    
    # Find repeated values (those with a count greater than 1)
    repeated_values = {val: count for val, count in value_counts.items() if count > 49}
    
    return repeated_values

# Find repeated values in column 1 for both data sets
repeated_SP = find_repeated_values(data_SP)
repeated_NSP = find_repeated_values(data_NSP)

# Print the repeated values in both data sets
print("Repeated values in data_SP (column 1):")
print(repeated_SP)

print("\nRepeated values in data_NSP (column 1):")
print(repeated_NSP)
```

```python
# Function to compare repeated values between the two datasets
def compare_repeated_values(repeated_SP, repeated_NSP):
    # Find common values between repeated_SP and repeated_NSP
    common_values = set(repeated_SP.keys()).intersection(repeated_NSP.keys())
    
    return common_values

# Compare repeated values between both datasets
common_repeated_values = compare_repeated_values(repeated_SP, repeated_NSP)

# Print the common repeated values
print("Common repeated values in column 1 between data_SP and data_NSP:")
print(common_repeated_values)
```
